### ** Model **

```python
data["Journey_day"] = data['Journey_Date'].str.split('/').str[0].astype(int)
data["Journey_month"] = data['Journey_Date'].str.split('/').str[1].astype(int)
data.drop(["Journey_Date"], axis = 1, inplace = True)
```

```python
data["Departure_hour"] = pd.to_datetime(data["Departure_Time"]).dt.hour
data["Departure_min"] = pd.to_datetime(data["Departure_Time"]).dt.minute
data.drop(["Departure_Time"], axis = 1, inplace = True)
```

```python
data["Arrival_hour"] = pd.to_datetime(data.Arrival_Time).dt.hour
data["Arrival_min"] = pd.to_datetime(data.Arrival_Time).dt.minute
data.drop(["Arrival_Time"], axis = 1, inplace = True)
```
